```python
import matplotlib.pyplot as plt
import numpy as np
import math
%pylab inline
```

```python
h=0.01
w=40
theta=[0.0]
velocidad=4
t=np.arange(0,10,0.1)
x=[0.0]
y=[0.0]
```

```python
for i in range(99):
    aceleracion = -w*theta[i]
    velocidad = velocidad + h*aceleracion
    theta.append(theta[i] + h*velocidad)
    x.append(0.225*theta[i+1])
    y.append(0.225 - 0.225*cos(theta[i+1]))
```

```python
np.asarray(theta)
y
```

```python
plt.plot(t,x)
```

```python
plt.plot(x,y)
```
